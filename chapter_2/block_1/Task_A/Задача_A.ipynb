```python
# 1)
import pandas as pd
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
```

```python
# 2)
df3 = pd.concat([df1, df2])
```

```python
X = df3.drop('Onshore/Offshore', axis=1)
y = df3['Onshore/Offshore']
```

```python
y
```

```python
# 3)
X_transform = pd.get_dummies(X) # Выполняем преобразование признаков 
X_transform
```

```python
# 4)
df = pd.concat([X_transform, y], axis = 1)
```

```python
# 5)
df_train = df.dropna() 
df_test = df[df['Onshore/Offshore'].isnull()] 
```

```python
df_test = df_test.drop('Onshore/Offshore', axis = 1)
df_test
```

```python
X1 = df_train.drop('Onshore/Offshore', axis=1)
y1 = df_train['Onshore/Offshore']
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=200)
clf.fit(X1_train, y1_train)
```

```python
y1_pred = clf.predict(X1_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y1_test, y1_pred)
```

```python
y2_pred = clf.predict(df_test)
y2_pred
```
