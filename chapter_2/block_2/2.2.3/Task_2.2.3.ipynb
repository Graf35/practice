```python
img = cv2.imread('Morph.jpg')

def draw_image3(image, imageNew):
    plt.figure(figsize=(16,13), dpi=90)
    plt.subplot2grid((3,3), (0,0))
    plt.imshow(image)
    plt.title('Исходное изображение')
    plt.subplot2grid((3,3), (0,1))
    plt.imshow(imageNew)
    plt.title('Преобразованное изображение')
    plt.subplot2grid((3,3), (0,2))
    plt.hist(np.ravel(image), bins=256, label=['Оригинал'])
    plt.hist(np.ravel(imageNew), bins=256, label=['Результат']) 
    plt.title('Гистограмма интенсивности цвета')
    plt.legend()
    plt.show()
```

```python
img_path = "morph3.bmp"
img = Image.open(img_path).convert('RGB')
```

```python
def draw_image_3(image, imageNew, n):
    plt.figure(figsize=(16,13), dpi=100)
    plt.subplot2grid((3,3), (0,0)) #1 - исходное изображение
    plt.title('Исходное изображение')
    plt.imshow(image) #Вывод
    plt.subplot2grid((3,3), (0,1)) #2 - преобразованное изображени
    plt.title('Преобразованное изображение')
    plt.imshow(imageNew) #Вывод
    #Гистограммы двух изображений
    plt.subplot2grid((3,3), (0,2)) 
    plt.hist(np.ravel(image), bins=256, label=['Оригинал'])
    plt.hist(np.ravel(imageNew), bins=256, label=['Результат']) 
    plt.legend()
    plt.title('n = %d' % n)
    plt.show()

def transform(f, img):
    img_copy = img.copy() # Преобразовывать будем копию исходного изображения
    draw = ImageDraw.Draw(img_copy) #Создаем инструмент для рисования
    width = img_copy.size[0] #Определяем ширину
    height = img_copy.size[1] #Определяем высоту 
    pixels = img_copy.load() #Выгружаем значения пикселей
    for i in range(width):
        for j in range(height):
            r, g, b = pixels[i, j]
            S = tuple(map(f, (r, g, b)))
            draw.point((i, j), S)    
    draw_image3(img, img_copy) 

```

```python
def f_bin(x):
    return 255 if x in range(200, 220) else 0

transform(f_bin, img)
```

```python
def f_bin(x):
    return 0 if x in range(0, 20) else 255

transform(f_bin, img)
```

```python
def f_bin(x):
    return 0 if x in range(40, 60) else 255

transform(f_bin, img)
```

```python
def f_bin(x):
    return 0 if x in range(90, 110) else 255

transform(f_bin, img)
```

```python
def SobelFilter1(image):
    SobelMtx = [[0,1],[-1,0]]# ядро фильтра Собеля
    kernel = np.array(SobelMtx, np.float32)# приведение ядра фильтра к типу float
    image1 = cv2.filter2D(image, -5, kernel)
    SobelMtx = np.matrix(SobelMtx).T
    kernel = np.array(SobelMtx, np.float32)# приведение ядра фильтра к типу float
    image2 = cv2.filter2D(image, -5, kernel)
    draw_image_3(image, cv2.add(image1, image2), 3)


img = cv2.imread('morph3.bmp')
SobelFilter1(img)
```

```python
def median_filter(image):
    parametr = 3 #Значения размера ядра
    #Применение фильтра для изображения
    draw_image_3(image, cv2.medianBlur(image, parametr), parametr)
    return cv2.medianBlur(image, parametr)

img = cv2.imread('Fingerprint.jpg')
image = median_filter(img)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (1,1))
transform_img = cv2.erode(image, kernel, iterations=1)
draw_image3(img, transform_img)
```
