### Пример построения дерева решений

В этом примере мы рассмотрим построение дерева решений, определяющее вид ириса по набору его характеристик.

**Ирисы Фишера** - самый популярный набор данных, используемый для иллюстрации работы различных алгоритмов классификации. 

Выборка состоит из 150 экземпляров ирисов трех видов (по 50 экземпляров каждого вида): Setosa, Versicolor, Virginica.   
Для каждого экземпляра ириса в данных содержатся четыре характеристики: длина и ширина чашелистика (sepal length и sepal width), длина и ширина лепестка (petal length и petal width).

![img](https://miro.medium.com/max/1000/1*lFC_U5j_Y8IXF4Ga87KNVg.png)

```python
from sklearn.datasets import load_iris # для загрузки набора данных
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree   # функция для отрисовки дерева
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Загружаем данные
iris = load_iris()
X, y = iris.data, iris.target
# X - признаки
# y - ответы 
```

```python
X
```

```python
y
```

```python
# Разобьем выборку на обучающую и тестовую
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

clf = DecisionTreeClassifier() # Создаем экземпляр класса
clf = clf.fit(X_train, y_train) # Обучаем модель

plt.figure(figsize=(20,10)) 
plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names) # Отобразим дерево
plt.show()

print('accuracy: ', clf.score(X_test, y_test)) # Оценим качество модели на тестовой выборке с помощью доли правильных ответов
```
